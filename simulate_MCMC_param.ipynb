```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# simulation parameters.
starting_T = [0.7, 0.1, 0.01]
mcmc_halflife = 1000
given_steps = range(5000)
delta = 0.10

# plot.
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5) )
ax1 , ax2 = axes.flatten()
for mcmc_t_acc_initial in starting_T:
    T_accs = mcmc_t_acc_initial * ( (np.exp(np.log(0.5) / mcmc_halflife))**given_steps)
    ax1.plot(given_steps, T_accs, lw=3)
    ax2.plot(given_steps, np.exp( -delta/T_accs  ), label="T_ini=%s"%mcmc_t_acc_initial , lw=3)
      
ax1.set_xlabel("step")
ax1.set_ylabel("T_acc")
ax2.set_ylabel("acceptance probability")
ax2.set_xlabel("step")
ax2.legend(loc = 'center right', fontsize=14, bbox_to_anchor=(1.60, 0.5))

plt.show()
```
